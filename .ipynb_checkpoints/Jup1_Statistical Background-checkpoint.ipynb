# Establishing Sample Statistics

This section establishes the initial information about the sample data, considering the visible differences between the forms of activity, and the distribution of weight, height, age and gender characteristics across the sample. 

Initially the data was imported, using the code below:

```python
#import programmes
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # making plots
```

The sample was initially investigated to consider any biases that it may create within the impeding machine learning

```python
path ="data_subjects_info.csv" #importing the data within a dataframe
Subject_Info = pd.read_csv(path)
```

The distribution of weight, height and age can be visualised below. Height forms a boxplot with a fairly central median, and relatively even whiskers, implying the distribution is not skewed. Weight is particularly concentrated within the lower and upper middle quartiles.Meanwhile, age is fairly evenly distributed with a noted outlier - all participants are under the age of 50.

```python
Subject_Info[['weight','height','age']].boxplot()
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
```

```python
ax = Subject_Info['gender'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Gender Distribution in the sample")
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
```

1 is male, 0 is female - noticably higher proportion of males than females within the sample. 

The known traits for each of the participants were then tested for correlation trends

```python
f = plt.figure(figsize=(10, 10))
plt.matshow(Subject_Info.corr(), fignum=f.number)
plt.xticks(range(Subject_Info.shape[1]), Subject_Info.columns, fontsize=14)
plt.yticks(range(Subject_Info.shape[1]), Subject_Info.columns, fontsize=14)
cb = plt.colorbar(fraction=0.03)
cb.ax.tick_params(labelsize=10)
plt.title('Correlation Matrix', fontsize=16);
```

As the chart above shows, Gender shows a positive correlation with weight and height (as a value of 0.4-0.8). The potential link between these physical features and gender implies that gender may be interpreted from recording of movement
